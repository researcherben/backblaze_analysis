```python
import pandas
print('pandas',pandas.__version__)
import glob
import pickle
import numpy
import time
import matplotlib.pyplot as plt
```

get the headers for the dataframe from a CSV

```python
df_header_only=pandas.read_csv('zipped_data/data_Q2_2018.zip_folder/2018-04-01.csv',nrows=3)
#df_header_only.head()
```

```python
df = pandas.read_csv('data_synthesized_from_csvs/all_failures_smart_columns1to31.log',header=None)
df.columns=df_header_only.columns[0:31]
df.shape
```

# histogram of power-on hours at time of failure

```python
n, bins, patches = plt.hist(df['smart_9_raw'].dropna(how='any'), bins=100, facecolor='blue', alpha=0.5)
plt.xlabel('power on hours',fontsize=12)
plt.ylabel('count',fontsize=12);
plt.title('all drives',fontsize=14);
```

### zoom in on relevant lifespans

```python
n, bins, patches = plt.hist(df[df['smart_9_raw'] < 50000]['smart_9_raw'].dropna(how='any'), bins=100, facecolor='blue', alpha=0.5)
plt.xlabel('power on hours',fontsize=12)
plt.ylabel('count',fontsize=12);
plt.title('all drives',fontsize=14);
```

```python
n, bins, patches = plt.hist(df[df['smart_9_raw'] < 50000]['smart_9_raw'].dropna(how='any')/24/30, bins=100, facecolor='blue', alpha=0.5)
plt.xlabel('power on months',fontsize=12)
plt.ylabel('count',fontsize=12);
plt.title('all drives',fontsize=14);
```

### Focus on most popular model, ST4000DM000

```python
n, bins, patches = plt.hist(df[df['model']=='ST4000DM000']['smart_9_raw'].dropna(how='any'), bins=100, facecolor='blue', alpha=0.5)
plt.xlabel('power on hours',fontsize=12)
plt.ylabel('count',fontsize=12);
plt.title('model ST4000DM000',fontsize=14);
```

```python
n, bins, patches = plt.hist(df[df['model']=='ST4000DM000']['smart_9_raw'].dropna(how='any')/24/30, bins=100, facecolor='blue', alpha=0.5)
plt.xlabel('power on months',fontsize=12)
plt.ylabel('count',fontsize=12);
plt.title('model ST4000DM000',fontsize=14);
```
