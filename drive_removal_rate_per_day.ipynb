```python
import pandas
print('pandas',pandas.__version__)
import glob
import pickle
import numpy
import datetime
import time
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('data_synthesized_from_csvs/failures_vs_drive_count_per_day.dat',delimiter=' ',header=None)
df.columns=['date in filename','number of drives removed from service','number of unique drives']
df.shape
```

```python
#df.head()
```

```python
df['date']=df['date in filename'].apply(lambda x: datetime.datetime.strptime(x.split('_')[-1],'%Y-%m-%d'))
```

```python
#df.head()
```

```python
df.drop(['date in filename'], axis=1,inplace=True)
```

```python
df['ratio of drives removed from service to count per day']=df['number of drives removed from service']/df['number of unique drives']
```

```python
#df.head()
```

```python
plt.plot_date(x=df['date'],y=df['number of unique drives'])
plt.xlabel('date',fontsize=12)
plt.ylabel('number of unique drives',fontsize=12);
plt.title('Backblaze drives',fontsize=12);
#plt.xticks(rotation=70);
```

```python
plt.plot_date(x=df['date'],y=df['number of drives removed from service'],markersize=3)
plt.xlabel('date',fontsize=12)
plt.ylabel('number of drives removed from service',fontsize=12);
plt.title('Backblaze drives',fontsize=12);
```

```python
plt.plot_date(x=df['date'],y=df['ratio of drives removed from service to count per day'],markersize=3)
plt.xlabel('date',fontsize=12)
plt.ylabel('(removed drive count)/(drive count per day)',fontsize=12);
plt.title('Backblaze drives',fontsize=12);
```
